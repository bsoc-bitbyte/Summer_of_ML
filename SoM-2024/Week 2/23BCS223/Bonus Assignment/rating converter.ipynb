Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

Loading the dataframe


```python
df=pd.read_csv("final_cccf.csv")
df.shape
```

```python
df.head(5)
```

```python
x=df['cc_rating']
y=df['cf_rating']
x=x.dropna()
y=y.dropna()
print(len(x),len(y))
```

```python
df1=df.dropna()
x=df1['cc_rating'].values
y=df1['cf_rating'].values
```

```python
df1
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot CC vs CF")
plt.show
```

```python
mean_x=x.mean()
mean_y=y.mean()
m=len(x)
numer = 0
denom = 0
for i in range(m):
  numer += (x[i] - mean_x) * (y[i] - mean_y)
  denom += (x[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)
print(m,c)
```

```python
rangex=np.linspace(x.min(),x.max(),num=200)
predicted_cf=(m*rangex)+c
plt.plot(rangex,predicted_cf,color='black',label="Regression Line")
plt.scatter(x,y,label="Actual data points")
plt.ylabel("Codeforces Rating")
plt.xlabel("Codechef Rating")
plt.legend()
plt.show()
```

```python
def predict(codechef_rating):
    codeforces_rating=(codechef_rating*m)+c
    return codeforces_rating
    
for a in range(700,2100,20):
    print("Codechef :",a,"Codeforces :",int(predict(a)))
```

```python
#linear Regression Model

Y=df_1['cf_rating'].values
X=df_1['cc_rating'].values
x_mean=np.mean(X)
y_mean=np.mean(Y)
num=0
den=0
for i in range(len(X)):
  num+=(X[i]-x_mean)*(Y[i]-y_mean)
  den+=(X[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
y_pred=b0+b1*x


```

```python
#to print the values of codechef and codeforces
#slope=b1
#intercpt=b0
def fun(codechef_rating):
    codeforces_rating=(codechef_rating*b1)+b0
    return codeforces_rating

for z in range(700,4100,50):
    print("Codechef :",z,"Codeforces :",round(fun(z),2))
```

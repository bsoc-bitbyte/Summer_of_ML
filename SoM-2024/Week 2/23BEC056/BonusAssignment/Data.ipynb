```python
filteredData = data.dropna(subset=['cf_rating','cc_rating'])
```

```python
filteredData = filteredData.astype({"cf_rating" : int, "cc_rating" : int})
filteredData.dtypes
```

```python
filteredData
```

```python
filteredData.plot.scatter(x = "cf_rating", y = "cc_rating", color = "black", alpha =0.3)
```

```python
def linearModel_ccratings(weight, intercept,cfRating):
  ccRating = weight*cfRating + intercept
  return ccRating
```

```python
import numpy as np

weight1 = 0.79
intercept1 = 575

cfRating_range = np.linspace(filteredData["cf_rating"].min(), filteredData["cf_rating"].max(), num = 6440)
predicted_ccRatings = linearModel_ccratings(weight1, intercept1, cfRating_range)
```

```python
filteredData.plot.scatter(x = "cf_rating", y = "cc_rating", color = "black", alpha =0.5)
plt.plot(cfRating_range, predicted_ccRatings, color = "blue")
```

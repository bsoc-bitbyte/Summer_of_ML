### Split data in X and Y

```python
x=np.array(df1['cc_rating'])
y=np.array(df1['cf_rating'])
```

### Visualize the data

```python
plt.scatter(x,y)
```

## using denominator method to find weight(w) and Bias(b) in y_pred=w*x+b

```python
denominator=x.dot(x)-x.mean()*x.sum()
w=(x.dot(y)-y.mean()*x.sum())/denominator
b=(x.dot(x)*y.mean()-x.mean()*x.dot(y))/denominator
w,b
```

### Predicting values equation

```python
# codechef to codeforce  y_pred-->codeforce rating & x-->codechef rating
y_pred=w*x+b
```

```python
#codeforce to codechef  y-->codeforce rating & x_predict-->codechef rating
# y=2345
# x_predict=(y-b)/w
```

### Visualizing the line and scatter plot of linear regression eqn

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.show()
```

### R_Squared Error

```python
d1=y-y_pred
d2=y-y.mean()
r2=1-d1.dot(d1)/d2.dot(d2)
r2
```

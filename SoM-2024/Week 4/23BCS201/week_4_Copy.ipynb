## encoding by dummy variable method

````python
one_hot_encoded_df = pd.get_dummies(x, columns=['Sex','Embarked','title'])
one_hot_encoded_df.head()
#convert boolen entry into binary
````

````python
#convert boolen entry into binary
one_hot_encoded_df['Sex_female']=one_hot_encoded_df['Sex_female'].astype(int)
one_hot_encoded_df['Sex_male']=one_hot_encoded_df['Sex_male'].astype(int)
one_hot_encoded_df['Embarked_C']=one_hot_encoded_df['Embarked_C'].astype(int)
one_hot_encoded_df['Embarked_Q']=one_hot_encoded_df['Embarked_Q'].astype(int)
one_hot_encoded_df['Embarked_S']=one_hot_encoded_df['Embarked_S'].astype(int)
one_hot_encoded_df['title_Master']=one_hot_encoded_df['title_Master'].astype(int)
one_hot_encoded_df['title_Miss']=one_hot_encoded_df['title_Miss'].astype(int)
one_hot_encoded_df['title_Mr']=one_hot_encoded_df['title_Mr'].astype(int)
one_hot_encoded_df['title_Mrs']=one_hot_encoded_df['title_Mrs'].astype(int)
one_hot_encoded_df['title_Others']=one_hot_encoded_df['title_Others'].astype(int)


````

````python
one_hot_encoded_df.head()
````

## Scaling done By Using Standard Scaler

````python
#scaling
scaler=StandardScaler()
x=scaler.fit_transform(one_hot_encoded_df)
````

````python
scaled_x=pd.DataFrame(x,columns=one_hot_encoded_df.columns)
scaled_x
````

````python
scaled_x.isnull().sum()
````

````python
#train_test split
X_train,X_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=42)
````

## **Random Forest**

````python
rf=RandomForestClassifier(n_estimators=30)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
````

````python
# Random Forest score
accuracy_score(y_test,y_pred)
````

### cross validation

````python
cv_rf=cross_val_score(rf,scaled_x,y,cv=15)
cv_rf .mean()
````

### Hypertuning parameter

````python
# parameter grid
param_grid_rf = {
    'n_estimators': [10, 30, 50],
    'max_depth': [None, 10, 20, 30],
}
````

````python
# Initialize GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf,
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
````

````python
# Fit GridSearchCV
grid_search_rf.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search_rf.best_params_
print(f'Best parameters: {best_params}')


````

````python
# Train the model with  best parameters
best_rf = grid_search_rf.best_estimator_
best_rf.fit(X_train, y_train)
# Make predictions
y_pred = best_rf.predict(X_test)


````

````python

# Evaluate the model
score = accuracy_score(y_test, y_pred)
print(f'Model accuracy: {score}')
````

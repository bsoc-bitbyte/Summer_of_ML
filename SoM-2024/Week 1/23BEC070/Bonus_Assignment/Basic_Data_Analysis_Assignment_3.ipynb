### How many unique countries are there in western europe, display them all.

### Display the data of top 5 east asian countries based on 'Generosity'

```python
df3 = df1[df1['Regional indicator'] == 'East Asia']
df4=df3.head(5)
sns.lineplot(data=df4 , x='Country name', y='Generosity')#9
plt.grid()
```

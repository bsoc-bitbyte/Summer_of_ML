### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
sns.relplot(data=df,x=df['Attack'],y=df['Defense'],col='Legendary',hue='Generation')
```

- #### From the graph we can clearly say that there is a wide range of non Legendary pokemon as compare to Legendary ones.
- #### Also Legendary Pokemon generally have higher Attack and Defense stats compared to non-Legendary Pokemon. 

```python
sns.kdeplot(df.Generation,fill=True,color='magenta')
```

- #### It is a density plot showing Pokemon distribution across different generations
- #### In the graph Peaks can be seen around the generation 1, 3 and 5 indicating that higher no. of pokemons are from these Generations while there are relatively less pokemon from 2nd and 4th Generations

```python
sns.kdeplot(df.Total,fill=True,color='navy')
```

- #### It is a density plot showing pokemon distribution based on their total stats.
- #### We can see two peaks around 300-400 and 500-600 indicating higher no. of Pokemons have their total stats in these ranges.Most common range being the 500-600.
- #### The significant decrease in graph after 600 shows that very few pokemons have total stats higher than 600.

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
df3=df[df.Legendary==False][df.Generation==4][df.Overall_Type.str.contains("Ghost")]
df4=df3.sort_values(by=['Speed'],ascending=False)
print(df4.iloc[1,1])
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
charizard_def=df.loc[df['Name']=='Charizard','Defense'].values[0]
charizard_atk=df.loc[df['Name']=='Charizard','Attack'].values[0]
print("ANSWER:",df[df.Legendary==False][df.Defense>charizard_def][df.Attack<charizard_atk]['Name'].count())
```

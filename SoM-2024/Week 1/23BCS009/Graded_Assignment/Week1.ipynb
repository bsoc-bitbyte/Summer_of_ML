### Which type of legendary pokemons are the most common?

```python
# total legendary
legendary_pokemon = df[df['Legendary'] == True]

# Count the occurrences of each type
type_counts = legendary_pokemon['Type 1'].value_counts()
```

```python
# Find the most common type
most_common_type = type_counts.idxmax()
most_common_count = type_counts.max()
```

```python
print(f"The most common type of legendary Pokémon is: {most_common_type} with {most_common_count} occurrences")
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
# Filter the dataset for Generation 6 legendary Pokémon
gen6_legendary_pokemon = df[(df['Generation'] == 6) & (df['Legendary'] == True)]

# Find the lowest HP among Generation 6 legendary Pokémon
weakest_gen6_legendary_hp = gen6_legendary_pokemon['HP'].min()

# Filter the dataset for Generation 3 non-legendary Pokémon
gen3_non_legendary_pokemon = df[(df['Generation'] == 3) & (df['Legendary'] == False)]

# Count how many Generation 3 non-legendary Pokémon have higher HP than the weakest Generation 6 legendary Pokémon
higher_hp_count = (gen3_non_legendary_pokemon['HP'] > weakest_gen6_legendary_hp).sum()

print(f"Number of Generation 3 non-legendary Pokémon with higher HP than the weakest Generation 6 legendary Pokémon: {higher_hp_count}")
```

### Print out the third slowest pokemon(s) in the dataset.

```python
# Sort the dataset based on the 'Speed' column in ascending order
sorted_pokemon_df = df.sort_values(by='Speed')

# Select the third entry from the bottom of the sorted dataset
third_slowest_pokemon = sorted_pokemon_df.iloc[-3:]

print("The third slowest Pokémon(s) in the dataset is/are:")
print(third_slowest_pokemon)
```

### Which pokemon type has the highest average HP?

```python
# Group the dataset by 'Type 1' and calculate the average HP for each type
average_hp_by_type = df.groupby('Type 1')['HP'].mean()

# Find the Pokémon type with the highest average HP
highest_avg_hp_type = average_hp_by_type.idxmax()
highest_avg_hp_value = average_hp_by_type.max()

print(f"The Pokémon type with the highest average HP is '{highest_avg_hp_type}' with an average HP of {highest_avg_hp_value:.2f}.")
```

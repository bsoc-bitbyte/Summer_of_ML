**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
country = df1[df1["Regional indicator"]=="East Asia"][["Country name","Generosity"]].sort_values("Generosity",ascending=False).head()

x=country["Country name"]
y=country["Generosity"]

plt.figure(figsize=(12,5)) 
plt.xlabel("Country")
plt.ylabel("Generosity")
plt.title("East Asian countries with top Generosity index")

indexes=np.arange(5)
plt.xticks(indexes,x)
plt.bar(x,y,width=0.4)
plt.grid()
plt.legend(["bar"])
```

```python
country = df1[df1["Regional indicator"]=="East Asia"][["Country name","Generosity","Ladder score"]].sort_values("Generosity",ascending=False).head()
country
```

```python
x=country["Country name"]
y=country["Ladder score"]

plt.figure(figsize=(12,5)) 

plt.plot(x,y,"--ro")
plt.xticks(x)
plt.xlabel("Country")
plt.ylabel("Ladder Score")
plt.title("Ladder Scores of Top East Asian Countries")
plt.legend(["line"])
plt.grid()
```

```python
country = df1[df1["Regional indicator"]=="Southeast Asia"][["Country name","Generosity","Ladder score"]].sort_values("Generosity",ascending=False).head()
country
```

```python
x=country["Country name"]
y=country["Ladder score"] 

plt.plot(x,y,"--ro")
plt.xticks(x)
plt.xlabel("Country")
plt.ylabel("Ladder Score")
plt.title("Ladder Scores of Top South-East Asian Countries")
plt.legend(["line"])
plt.grid()
```

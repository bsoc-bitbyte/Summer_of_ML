# **Visualization using Matplotlib Library**

## 10.Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)


```python
arr=df1[df1['Regional indicator']=='East Asia'].Generosity.sort_values(ascending=False).keys()[:5]
# print(arr)

new_df=df1[df1.index.isin(arr)]
plt.plot(new_df['Generosity'] , new_df['Ladder score'] )
plt.xlabel('Generosity')
plt.ylabel('Ladder score')
plt.title('Ladder score vs Generosity(East Asia)')
plt.grid(True)
plt.show()
```

## 11.Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries

```python
arr=df1[df1['Regional indicator']=='Southeast Asia'].Generosity.sort_values(ascending=False).keys()[:5]
# print(arr)

new_df=df1[df1.index.isin(arr)]
plt.plot(new_df['Generosity'] , new_df['Ladder score'] )
plt.xlabel('Generosity')
plt.ylabel('Ladder score')
plt.title('Ladder score vs Generosity(Southeast Asia)')
plt.grid(True)
plt.show()
```

## 12.Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal','Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'

```python
countries=['China', 'Nepal','Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']

new_df =df1[df1['Country name'].isin(countries)]

plt.scatter(new_df['Country name'] , new_df['Logged GDP per capita'] , marker='x' , c='green')

plt.xlabel('Country')
plt.ylabel('Logged GDP per capita')
plt.title('Logged GDP per capita vs Country')
plt.grid(True)
plt.show()
```

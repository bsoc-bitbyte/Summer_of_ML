# **Visualization of results using Matplotlib library**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/world-happiness-report-2021.csv')
```

```python
x = df1[df1["Regional indicator"] == "East Asia"].sort_values(by="Generosity", ascending=False).head(5)
print("The data of top 5 countries :")
x
```

```python
x = df1[df1["Regional indicator"] == "East Asia"].sort_values(by="Generosity", ascending=False).head(5)["Country name"]
y = df1[df1["Regional indicator"] == "East Asia"].sort_values(by="Generosity", ascending=False).head(5)["Ladder score"]
plt.figure(figsize=(15,10))
plt.plot(x, y)
plt.xlabel("Country name")
plt.ylabel("Ladder score")
plt.show()
```

```python
x = df1[df1["Regional indicator"] == "Southeast Asia"].sort_values(by="Ladder score", ascending=False).head(5)["Country name"]
y = df1[df1["Regional indicator"] == "Southeast Asia"].sort_values(by="Ladder score", ascending=False).head(5)["Ladder score"]
plt.figure(figsize=(15,10))
plt.plot(x, y)
plt.xlabel("Country name")
plt.ylabel("Ladder score")
plt.show()
```

```python
df_2021 = df1[df1["Country name"].isin(['China', 'Nepal',
'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'])]

x = df_2021["Country name"]
y = df_2021["Logged GDP per capita"]

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.xlabel("Country name")
plt.ylabel("Logged GDP per capita")
plt.show()
```

***1. Display the data of top 5 east asian countries based on 'Generosity'***

```python
east_asian_region = "East Asia"

# Filter the dataset for East Asian countries
east_asian_df = df1[df1['Regional indicator'] == east_asian_region]

# Sort the filtered dataset based on 'Generosity' in descending order and select the top 5
top_5_east_asian_generosity = east_asian_df.sort_values(by='Generosity', ascending=False).head(5)

# Display the top 5 East Asian countries based on Generosity
print("Top 5 East Asian countries based on Generosity:")
print(top_5_east_asian_generosity[['Country name', 'Generosity']])
```

***3. Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'***

```python
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']

# Create the DataFrame object df_2021
df_2021 = pd.DataFrame({'Country': countries})


# Merge df_2021 with df on 'Country name' column to include 'Logged GDP per capita'
df_2021 = pd.merge(df_2021, df1[['Country name', 'Logged GDP per capita']], how='left', left_on='Country', right_on='Country name')

# Drop the 'Country name' column as it is no longer needed
df_2021.drop(columns=['Country name'], inplace=True)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_2021, x='Country', y='Logged GDP per capita', color='blue')
plt.title('Relation between Countries and Logged GDP per capita')
plt.xlabel('Country')
plt.ylabel('Logged GDP per capita')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

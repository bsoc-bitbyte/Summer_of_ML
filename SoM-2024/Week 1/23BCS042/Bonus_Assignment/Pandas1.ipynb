### 6) What is the number of columns in the dataset, also print the name of all the columns?

### 8) Print only the occupation column

### 9) How many different occupations are in this dataset?


```python
#your code here
unique_occupations = users['occupation'].unique()
num_unique_occupations = len(unique_occupations)
print(num_unique_occupations)
```

### 10) What is the most frequent occupation?

```python
#your code here
most_frequent_occupation = users['occupation'].value_counts().idxmax()
print(most_frequent_occupation)
```

### 14) What is the age with least occurrence?

```python
#your code here
least_occurrence_age = users['age'].value_counts().idxmin()

print(f"The age with the least occurrence is: {least_occurrence_age}")
```

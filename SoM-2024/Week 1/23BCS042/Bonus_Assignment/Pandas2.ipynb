### 7) Which was the most-ordered item?

```python
#your code here
item_orders = Chipo.groupby('item_name')['quantity'].sum()

most_ordered_item = item_orders.idxmax()
print(most_ordered_item)
```

### 8) For the most-ordered item, how many items were ordered?

```python
#your code here
most_ordered_quantity = item_orders.max()
print(most_ordered_quantity)
```

### 9) What was the most ordered item in the choice_description column?

```python
#your code here
choice_orders = Chipo.groupby('choice_description')['quantity'].sum()
most_ordered_choice = choice_orders.idxmax()
print(most_ordered_choice)
```

### 10) How many items were ordered in total?

```python
#your code here
total_items_ordered = Chipo['quantity'].sum()
print(total_items_ordered)
```

### 11) Turn the item price into a float

```python

Chipo['item_price'] = pd.to_numeric(Chipo['item_price'], errors='coerce')
```

### 12) How much was the revenue for the period in the dataset?
#### Revenue will be the sum of quantity * item_price!

```python
#your code here
Chipo['revenue'] = Chipo['quantity'] * Chipo['item_price']
total_revenue = Chipo['revenue'].sum()
print(total_revenue)
```

### 13) How many orders were made in the period?

```python
#your code here
num_orders = Chipo['order_id'].nunique()
print(num_orders)
```

### 14) What is the average revenue amount per order?

```python
#your code here
Chipo['revenue'] = Chipo['quantity'] * Chipo['item_price']

# Calculate the total revenue
total_revenue = Chipo['revenue'].sum()

# Calculate the total number of unique orders
num_orders = Chipo['order_id'].nunique()

# Calculate the average revenue amount per order
average_revenue_per_order = total_revenue / num_orders
print("Average revenue amount per order:", average_revenue_per_order)
```

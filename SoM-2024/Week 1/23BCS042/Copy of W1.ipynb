### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
gen4_ghost_non_legendary = df[(df['Generation'] == 4) & (df['Type 1'] == 'Ghost') & (df['Legendary'] == False)]

# Step 2: Sort the filtered DataFrame by Speed in descending order to find the fastest ones
sorted_ghosts_by_speed = gen4_ghost_non_legendary.sort_values(by='Speed', ascending=False)

# Step 3: Select the second fastest Pokemon
second_fastest_ghost = sorted_ghosts_by_speed.iloc[1]

# Print the name of the second fastest non-legendary 'Ghost' type Pokemon from 4th generation
print("The second fastest non-legendary 'Ghost' type Pokemon from 4th generation is:", second_fastest_ghost['Name'])
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
charizard = df[df['Name'] == 'Charizard'].iloc[0]
charizard_attack = charizard['Attack']
charizard_defense = charizard['Defense']

# Step 2: Filter for non-legendary Pokemon
non_legendary = df[df['Legendary'] == False]

# Step 3: Count non-legendary Pokemon with stronger Defense but weaker Attack than Charizard
stronger_defense_weaker_attack = non_legendary[(non_legendary['Defense'] > charizard_defense) & (non_legendary['Attack'] < charizard_attack)]

# Get the count of such PokÃ©eon
count_stronger_defense_weaker_attack = stronger_defense_weaker_attack.shape[0]

print("Number of non-legendary Pokmeon with stronger defense but weaker attack than Charizard:", count_stronger_defense_weaker_attack)
```

### Which type of legendary pokemons are the most common?

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
gen3_non_legendary_max_hp = df[(df['Generation'] == 3) & (~df['Legendary'])]['HP'].max()

# Step 2: Find the minimum HP of Generation 6 legendary Pokemon
gen6_legendary_min_hp = df[(df['Generation'] == 6) & (df['Legendary'])]['HP'].min()

# Step 3: Compare the maximum HP of Generation 3 non-legendary Pokemon with the minimum HP of Generation 6 legendary Pokemon
count_higher_hp_pokemon = df[(df['Generation'] == 3) & (~df['Legendary']) & (df['HP'] > gen6_legendary_min_hp)]['Name'].count()

print("Number of Generation 3 non-legendary Pokemon with higher HP than the weakest Generation 6 legendary Pokemon:", count_higher_hp_pokemon)

```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
filtered_df1 = poke1[(poke1['Legendary'] == False) & (poke1['Type 1'] == 'Ghost') & (poke1['Generation'] == 4)]

second_fast= filtered_df1['Speed'].nlargest(2).iloc[-1]

name_sec_fast = filtered_df1.loc[filtered_df1['Speed'] == second_fast, 'Name'].iloc[0]
print(f'The second fastest non-legendary "Ghost" type pokemon from 4th generation : {name_sec_fast}')
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
att_charizard = poke1.loc[poke1['Name'] == 'Charizard', 'Attack'].iloc[0]
def_charizard = poke1.loc[poke1['Name'] == 'Charizard', 'Defense'].iloc[0]
print(f'Attack and Defense of Charizard are "{att_charizard}" and "{def_charizard}" resp.')

filtered_df2 = poke1[(poke1['Legendary'] == False) & (poke1['Attack'] < att_charizard)  & (poke1['Defense'] > def_charizard)]
print(f'Number of non-legendary pokemons have stronger defence but weaker attack than Charizard : {len(filtered_df2)}')
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

### Which pokemon type has the highest average HP?

```python
avg_hp_by_type = poke1.groupby('Type 1')['HP'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_hp_by_type.index, y=avg_hp_by_type.values, palette='viridis')
plt.xlabel('Pokemon Type')
plt.ylabel('Average HP')
plt.title('Average HP by Pokemon Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_hp_by_type = poke1.groupby(['Type 1', 'Type 2'])['HP'].mean().unstack().fillna(0)

plt.figure(figsize=(10, 6))
sns.heatmap(avg_hp_by_type, cmap='viridis', annot=True, fmt=".0f", cbar_kws={'label': 'Average HP'})
plt.xlabel('Type 2')
plt.ylabel('Type 1')
plt.title('Average HP by Pokemon Types (Type 1 vs Type 2)')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

highest_hp_combination = avg_hp_by_type.stack().idxmax()
highest_avg_hp = avg_hp_by_type.stack().max()

print(f"The combination with the highest average HP is {highest_hp_combination} with an average HP of {highest_avg_hp}.")
```

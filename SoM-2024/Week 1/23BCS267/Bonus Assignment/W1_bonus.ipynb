## `Task 2 : Visualization of results using Matplotlib library`

### Display the data of top 5 east asian countries based on 'Generosity'

```python
top_5_east_asian_countries = df1[df1['Regional indicator'] == 'East Asia'].nlargest(5, 'Generosity')
top_5_east_asian_countries
```

### Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)

```python
plt.plot(top_5_east_asian_countries['Country name'], top_5_east_asian_countries['Ladder score'])
plt.xlabel('Country')
plt.ylabel('Ladder Score')
plt.title('Variation of "Ladder Score" among top 5 East Asian Countries')
plt.xticks(rotation=15)
plt.show()
```

### Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries

```python
south_east_asian_countries_ = df1[df1['Regional indicator'] == 'Southeast Asia']

south_east_asian_countries = south_east_asian_countries_['Country name'][:5]
south_east_asian_data = df1[df1['Country name'].isin(south_east_asian_countries)]

plt.plot(south_east_asian_data['Country name'], south_east_asian_data['Ladder score'])
plt.xlabel('Country')
plt.ylabel('Ladder Score')
plt.title('Variation of "Ladder Score" among 5 South East Asian Countries')
plt.show()
```

### Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'

```python
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(countries)]

plt.scatter(df_2021['Country name'], df_2021['Logged GDP per capita'])
plt.xlabel('Country')
plt.ylabel('Logged GDP per capita')
plt.title('Relation between Countries and Logged GDP per capita')
plt.show()
```

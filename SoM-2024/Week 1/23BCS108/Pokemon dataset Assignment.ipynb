### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
second_fastest_non_legendary_Ghost_type_pokemon = df[(df["Type_1"] == "Ghost") & (df['Legendary'] == 0) & (df['Generation'] == 4)].sort_values("Speed", ascending=False).iloc[1]
second_fastest_non_legendary_Ghost_type_pokemon['Name']
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
df.set_index("Name", inplace=True)

Attack = df["Attack"]["Charizard"]
Defense = df["Defense"]["Charizard"]

non_legendary_pokemons = df[(df["Legendary"] == 0) & (df["Attack"] < Attack) & (df["Defense"] > Defense)]
non_legendary_pokemons.shape[0]
```

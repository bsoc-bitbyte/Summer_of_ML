```python
western_europe_contries = df1[df1['Regional indicator'] == 'Western Europe']['Country name'].unique()
length = len(western_europe_contries)
length
```

```python
india_happiness = df1[df1['Country name'] == 'India'][['Country name','Ladder score']]
india_happiness
```

```python
east_asian_countries = df1[df1['Regional indicator'] == 'East Asia'].sort_values(by='Generosity', ascending = False).head(5)
east_asian_countries
```

```python
countries = east_asian_countries['Country name']
ladder_scores = east_asian_countries['Ladder score']
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(countries,ladder_scores,marker='o')
plt.title('Ladder Score Variation Among Top 5 East Asian Countries (Based on Generosity)')
plt.xlabel('Country')
plt.ylabel('Ladder Score')
plt.grid(True)
plt.show()
```

```python
countries_list = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(countries_list)]

plt.figure(figsize=(10, 6))
plt.scatter(df_2021['Country name'], df_2021['Logged GDP per capita'], color='b')
plt.title('Logged GDP per Capita for Selected Countries')
plt.xlabel('Country')
plt.ylabel('Logged GDP per Capita')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

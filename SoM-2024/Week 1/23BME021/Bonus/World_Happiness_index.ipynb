```python
a = (df1[(df1['Regional indicator'] == "East Asia")].sort_values(by="Generosity", ascending=False))[:5]

a.plot(kind="bar", x="Country name", y="Generosity")

plt.title("Top 5 Most Generous Countries")
plt.xlabel("Country")
plt.ylabel("Generosity Score")
plt.show()
```

```python
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
custom_df = df1[df1['Country name'].isin(countries)]
custom_df.plot(kind="line", x="Country name", y="Ladder score")

plt.title("Ladder score ")
plt.xlabel("country name")
plt.ylabel("ladder score")
plt.xticks(rotation=90)
plt.show()
```
